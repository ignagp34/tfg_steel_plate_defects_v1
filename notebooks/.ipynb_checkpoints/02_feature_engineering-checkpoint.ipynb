 ## RESULTADOS

A continuación se seleccionan las posibles transformaciones para cada atributo (Estas son una aproximación inicial tomando como referencia el skewness), se han descartado los targets y las columnas binarias. Sobre estas últimas también sería posible realizar un robust scaler.

```python
df_train = pd.read_csv(file_path)

feat_gen = FeatureGenerator(img_width=1700, img_height=12900000)
df_train_plus = feat_gen.fit_transform(df_train)


TARGET_COLS = [
    "Pastry", "Z_Scratch", "K_Scatch",
    "Stains", "Dirtiness", "Bumps", "Other_Faults"
]

ID_COLS = ["id"]                 # identificadores puros
# todo lo que tenga solo 2 valores únicos (0/1, True/False) lo tratamos como binario a estas no es necesario realizar ninguna transformación:

BINARY_COLS = (
    df_train_plus.nunique(dropna=False)
      .loc[lambda s: s == 2]
      .index.difference(TARGET_COLS + ID_COLS)
      .tolist()
)


NUMERIC_COLS = df_train_plus.columns.difference(
    TARGET_COLS + ID_COLS + BINARY_COLS
)
```

Seleccionamos transformación de la columna correspondiente según skewness:

```python
#skew/outliers sobre NUMERIC_COLS
SK_SKEW  = df_train_plus[NUMERIC_COLS].skew(numeric_only=True)
ABS_SKEW = SK_SKEW.abs()

def out_ratio(s):
    q1, q3 = s.quantile([0.25, 0.75])
    iqr = q3 - q1
    return ((s < q1 - 1.5*iqr) | (s > q3 + 1.5*iqr)).mean()

OUT_RATIO = df_train_plus[NUMERIC_COLS].apply(out_ratio)

LOG_COLS = [
    c for c in NUMERIC_COLS
    if df_train_plus[c].min() >= 0 and SK_SKEW[c] > 1.2
]
ROBUST_COLS = [
    c for c in NUMERIC_COLS
    if OUT_RATIO[c] > 0.02 and ABS_SKEW[c] <= 1.2
]
STD_COLS = [
    c for c in NUMERIC_COLS
    if OUT_RATIO[c] < 0.02 and ABS_SKEW[c] <= 0.5
]
QUANT_COLS = sorted(
    set(NUMERIC_COLS) - set(LOG_COLS) - set(ROBUST_COLS) - set(STD_COLS)
)

print("LOG_COLS: ",LOG_COLS) #  inicialmente  "Pixels_Areas","X_Perimeter","Y_Perimeter","Sum_of_Luminosity","Maximum_of_Luminosity",
print("QUANT_COLS: ",QUANT_COLS)
print("ROBUST_COLS: ",ROBUST_COLS)
print("STD_COLS",STD_COLS)
print("REDUNDANT_COLS: ",REDUNDANT_COLS)
```

## ENTRENAMIENTO
Importamos funciones creadas localmente para comparar. En el siguiente notebook se explicará con mayor detalle el funcionamiento del training.

```python
# --- Librerías de proyecto ---
from sklearn.preprocessing import FunctionTransformer   # identidad

from src.models.trainer import Trainer, TrainerConfig
from src.models.trainer import _compute_pos_weights      # helpers internos
from src.models.builders import build_estimator
```

Creamos Sub-clase que inhabilita el preprocesamiento pero reutiliza el resto de funciones de trainer.py

```python
class RawTrainer(Trainer):
    """Entrena exactamente igual que `Trainer` pero sin pipeline de features."""
    def _prepare_transformers_and_models(self, y_train):
        # 1) Preprocesador “identity” (pasa todo tal cual)
        self.preproc = FunctionTransformer(validate=False)
        # 2) Estimadores one-vs-rest con los mismos pesos
        pos_weights = _compute_pos_weights(y_train)
        self.classifiers = build_estimator(self.cfg.model_name, pos_weights)
```

Para los modelos lineales como logreg, sciklt envía warnings advirtiendo de que escalemos los features o aumentemos el número de iteraciones

```python
# Modelos que quieras probar (deben estar soportados por build_estimator)
MODELS = ["lgbm", "rf", "lr", "mlp"]       # añade/elimina a tu gusto "lgbm, rf, lr, mlp"
N_SPLITS = 2                       # mientras mayor sea el número de splits, mayor tiempo tomará.
RANDOM_STATE = 42

results = []

for model in MODELS:
    # ---------- 1) *con* preprocesamiento -------------------
    cfg = TrainerConfig(n_splits=N_SPLITS,
                        model_name=model,
                        random_state=RANDOM_STATE)
    trainer = Trainer(cfg)
    cv_df = trainer.run_cv()                    
    results.append({
        "model": model,
        "preproc": "sí",
        "mean_auc": cv_df["mean_auc"].mean()
    })
    
    #---------- 2) *sin* preprocesamiento -------------------
    cfg_raw = TrainerConfig(n_splits=N_SPLITS,
                            model_name=model,
                            random_state=RANDOM_STATE)
    raw_trainer = RawTrainer(cfg_raw)
    cv_raw_df = raw_trainer.run_cv()
    results.append({
        "model": model,
        "preproc": "no",
        "mean_auc": cv_raw_df["mean_auc"].mean()
    })

# -------------------------------------------------------------------------
# Tabla resumen
# -------------------------------------------------------------------------
results_df = pd.DataFrame(results)\
                .sort_values(["model", "preproc"])
display(results_df)
```

### Conclusiones
- Sin las transformaciones los modelos lineales no convergen y empeoran significativamente.
- Para modelos basados en arboles como random forest y lightboost el preprocesamiento no es necesario.
- El orden de transformaciones (winsorizar → log → scaler) reduce asimetrías y colas extremas.
- Las features derivadas añaden información clave (tamaño relativo, posición y luminosidad normalizada).
- El pipeline resultante es reproducible y encaja directamente en el modelo, evitando el data leakage.
